```python
# Recomendador de películas -> predecir que películas pueden tener éxito
# Importar archivos
import pandas as pd

movies = pd.read_csv("https://raw.githubusercontent.com/4GeeksAcademy/k-nearest-neighbors-project-tutorial/main/tmdb_5000_movies.csv")
credits = pd.read_csv("https://raw.githubusercontent.com/4GeeksAcademy/k-nearest-neighbors-project-tutorial/main/tmdb_5000_credits.csv")
movies.head()
```

```python
# Creación de una base de datos
import sqlite3

con = sqlite3.connect("../data/base_de_datos_peliculas.db")

movies.to_sql("movies", con, if_exists = "replace", index = False)
credits.to_sql("credits", con, if_exists = "replace", index = False)
```

```python
# Unión de ambas tablas
tabla_total = """
    SELECT *
    FROM movies
    INNER JOIN credits
    ON movies.title = credits.title;
"""
total_data = pd.read_sql_query(tabla_total, con)
total_data.head()
```

```python
# Limpieza de tablas para dejar sólo las columnas que indican en el enunciado
total_data = total_data.drop(["budget","id","homepage","original_language","original_title","popularity","production_companies","production_countries","release_date","revenue","runtime","spoken_languages","status","tagline","vote_average","vote_count"], axis = 1)

total_data.head()
```

```python
# Eliminación de duplicados (la columna 'title' está repetida) -> no funciona 
if total_data.duplicated().sum():
    total_data = total_data.drop_duplicates()
print(total_data.shape)
total_data.head()
```

```python
total_data.columns
```

```python
# Renombro la columna repetida para poder eliminarla, para que finalmente tenga las 7 columnas correspondientes
total_data.columns = ['genres', 'keywords', 'overview', 'title', 'movie_id', 'title2', 'cast','crew']
total_data.drop(columns=['title2'], inplace=True)
print(total_data.shape)
total_data.head()
```

```python
# Transformar los datos
import json

def load_json_safe(json_str, default_value = None):
    try:
        return json.loads(json_str)
    except (TypeError, json.JSONDecodeError):
        return default_value

# Seleccionar el atributo `name` y reemplazar las columnas `genres` y `keywords`   
total_data["genres"] = total_data["genres"].apply(lambda x: [item["name"] for item in json.loads(x)] if pd.notna(x) else None)
total_data["keywords"] = total_data["keywords"].apply(lambda x: [item["name"] for item in json.loads(x)] if pd.notna(x) else None)

# Seleccionar los tres primeros nombres de 'cast'
total_data["cast"] = total_data["cast"].apply(lambda x: [item["name"] for item in json.loads(x)][:3] if pd.notna(x) else None)

# Transformar 'crew' para que solo contenga el nombre del director
total_data["crew"] = total_data["crew"].apply(lambda x: " ".join([crew_member['name'] for crew_member in load_json_safe(x) if crew_member['job'] == 'Director']))

# Convertir 'overview' en una lista
total_data["overview"] = total_data["overview"].apply(lambda x: [x])

total_data.head(20)
```

```python
# Quitar espacios entre las palabras de las columnas `genres`, `cast`, `crew` y `keywords`
# Reducir las columnas convertidas anteriores en una sola columna llamada 'tags'ArithmeticError

total_data["overview"] = total_data["overview"].apply(lambda x: [str(x)])
total_data["genres"] = total_data["genres"].apply(lambda x: [str(genre) for genre in x])
total_data["keywords"] = total_data["keywords"].apply(lambda x: [str(keyword) for keyword in x])
total_data["cast"] = total_data["cast"].apply(lambda x: [str(actor) for actor in x])
total_data["crew"] = total_data["crew"].apply(lambda x: [str(crew_member) for crew_member in x])

total_data["tags"] = total_data["overview"] + total_data["genres"] + total_data["keywords"] + total_data["cast"] + total_data["crew"]
total_data["tags"] = total_data["tags"].apply(lambda x: ",".join(x).replace(",", " "))

total_data.drop(columns = ["genres", "keywords", "cast", "crew", "overview"], inplace = True)
```

```python
# Construir un KKN
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import NearestNeighbors

# Vectorizar el texto
vectorizer = TfidfVectorizer()
columna_vectorizada = vectorizer.fit_transform(total_data["tags"])

# Usamos distancia coseno
model = NearestNeighbors(n_neighbors = 6, algorithm = "brute", metric = "cosine")
model.fit(columna_vectorizada)

def get_movie_recommendations(movie_title):
    movie_index = total_data[total_data["title"] == movie_title].index[0]
    distances, indices = model.kneighbors(columna_vectorizada[movie_index])
    similar_movies = [(total_data["title"][i], distances[0][j]) for j, i in enumerate(indices[0])]
    return similar_movies[1:]

input_movie = "Harry Potter and the Half-Blood Prince"
recommendations = get_movie_recommendations(input_movie)
print("Film recommendations '{}'".format(input_movie))
for movie, distance in recommendations:
    print("- Film: {}".format(movie))
```
